```python
from PIL import Image
img=Image.open("/content/mam.jpeg")
img
```

```python
img1=img.convert("L")
img1
```

```python
import matplotlib.pyplot as plt
```

```python
traspose1=img.transpose(Image.FLIP_LEFT_RIGHT)
traspose1
```

```python
traspose2=img.transpose(Image.FLIP_TOP_BOTTOM)
traspose2
```

```python
rotate1=img.transpose(Image.ROTATE_90)
rotate1
```

```python
plt.subplot(1,6,1)
plt.imshow(img)
plt.title("Orginal Image")
plt.subplot(1,6,2)
plt.imshow(img1)
plt.title("Grayscale Image")
plt.subplot(1,6,3)
plt.imshow(traspose1)
plt.title("flip left to right")
plt.subplot(1,6,4)
plt.imshow(traspose2)
plt.title("flip top to bottom")
plt.subplot(1,6,5)
plt.imshow(rotate1)
plt.title("rotate Picture")
```
